## Переводим валюты в рубли для построения графиков 

```python
usd_mark_dollar = df['currency'] == '$'
usd_mark_euro = df['currency'] == '€'

df.loc[usd_mark_dollar, 'upper_salary'] = df.loc[usd_mark_dollar, 'upper_salary'] * 80 
df.loc[usd_mark_dollar, 'lower_salary'] = df.loc[usd_mark_dollar, 'lower_salary'] * 80

df.loc[usd_mark_euro, 'upper_salary'] = df.loc[usd_mark_euro, 'upper_salary'] * 93 
df.loc[usd_mark_euro, 'lower_salary'] = df.loc[usd_mark_euro, 'lower_salary'] * 93 

df.loc[usd_mark_dollar | usd_mark_euro, 'currency'] = '₽'
```

## Разворачиваем массив с навыками 

```python
dictionary_popular = {}

for sk_list in df['tags']:
    sk_list = sk_list.strip('[]')
    items = sk_list.split(',')
    items = [item.strip() for item in items]
    items = [item.strip('\'') for item in items]
    for item in items: 
        if item not in dictionary_popular: 
            dictionary_popular[item] = 1 
        else: 
            dictionary_popular[item] += 1
```

## Формируем топ навыков по всем вакансиям  из словаря 

```python
total_skills_df = pd.DataFrame(dictionary_popular.items(), columns=['skill', 'count'])

total_skills_df = total_skills_df.sort_values('count', ascending=False).reset_index(drop=True)

total_skills_df.head(10)
```

## Самые популярные навыки для вакансий без опыта

```python
skills = {}

for skills_choice in df['tags'][df['experience'] == 'Требуемый опыт работы: не требуется']:
        skills_choice = skills_choice.strip('[]')
        items = skills_choice.split(',')
        items = [item.strip() for item in items]
        items = [item.strip('\'') for item in items]
        for skill in items:
            if skill in skills:
                skills[skill] += 1
            else:
                skills[skill] = 1

skills_df = pd.DataFrame.from_dict(skills, orient='index', columns=['count'])
skills_df = skills_df.sort_values(by='count', ascending=False)
skills_df = skills_df[skills_df.index != ' ']

plt.figure(figsize=(12, 6)) 
sns.barplot(x=skills_df.index, y=skills_df['count'])
plt.xticks(rotation=45)
plt.xlabel('Skills')
plt.ylabel('Count')
plt.title('Most Frequently Occurring Skills')
plt.xlim(0, 20)  
plt.tight_layout()  
plt.show()
```

## Формируем топ самых популярных навыков для вакансий с опытом работы 1-3 года

```python
skills_1_3 = {}

for skills in df['tags'][df['experience'] == 'Требуемый опыт работы: 1–3 года']:
        skills = skills.strip('[]')
        items = skills.split(',')
        items = [item.strip() for item in items]
        items = [item.strip('\'') for item in items]
        for skill in items:
            if skill in skills_1_3:
                skills_1_3[skill] += 1
            else:
                skills_1_3[skill] = 1

skills_df_1_3 = pd.DataFrame.from_dict(skills_1_3, orient='index', columns=['count'])
skills_df_1_3 = skills_df.sort_values(by='count', ascending=False)
skills_df_1_3 = skills_df[skills_df.index != ' ']

plt.figure(figsize=(12, 6)) 
sns.barplot(x=skills_df_1_3.index, y=skills_df_1_3['count'])
plt.xticks(rotation=45)
plt.xlabel('Skills')
plt.ylabel('Count')
plt.title('Most Frequently Occurring Skills')
plt.xlim(0, 20)  
plt.tight_layout()  
plt.show()
```
